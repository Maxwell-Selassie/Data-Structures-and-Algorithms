```python
# two sum problem
from typing import List
def two_sum(nums: List[int], target: int) -> List[int]:
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i

nums = [2, 4, 32, 21, 32]
target = 36

print(two_sum(nums, target))
```

```python
# contains duplicate
from typing import List
def contains_duplicate(nums: List[int]) -> bool:
    duplicated = set()
    for i in range(len(nums)):
        if nums[i] in duplicated:
            return True
        else:
            duplicated.add(nums[i])
    return False

nums = [2,4,5,6,8]
print(contains_duplicate(nums))
```

```python
# valid anagram
from collections import Counter

def valid_anagram(string1: str, string2: str) -> bool:
    string_1_hash = dict(Counter(string1))
    string_2_hash = dict(Counter(string2))
    return string_1_hash == string_2_hash

string1 = 'racecar'
string2 = 'carrace'
print(valid_anagram(string1, string2))
```

```python
# group valid anagrams
from collections import defaultdict
from typing import List

def group_valid_anagram(strs: List):
    group_anagram = defaultdict(list)
    for word in strs:
        key = ''.join(sorted(word))
        group_anagram[key].append(word)

    return list(group_anagram.values())

strs = ["act","pots","tops","cat","stop","hat"]
print(group_valid_anagram(strs))
```
